```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\USER\\Downloads\\Data Projects\\Kaggle loan data\\Loan Payments data.csv')

df.head(5)
```

```python
df['Loan_ID'] = range(1, 501)
df.head(503)
```

```python
df['loan_status'] = df['loan_status'].str.lower()
df.head()
```

```python
df['past_due_days'] = df['past_due_days'].fillna('')
```

```python
df
```

```python
df = df.fillna('')
```

```python
df.head(393)
```

```python
df['education'] = df['education'].replace('Bechalor', 'Bachelor')
```

```python
df
```

```python
#Count of loans by status
loan_status_counts = df['loan_status'].value_counts()
loan_status_counts
```

```python
#Average age of customers by education level, #Converting the age values to two decimal places
average_age_by_education = df.groupby('education')['age'].mean().round(2)
average_age_by_education
```

```python
#Distribution of ages in different loan statuses
age_stats_by_status = df.groupby('loan_status')['age'].describe()
age_stats_by_status
```

```python
#Loan term analysis
loan_term_counts = df['terms'].value_counts()
loan_term_counts
```

```python
df
```

```python
df
```

```python
df
```

```python
df['effective_date'] = pd.to_datetime(df['effective_date'])
df
```

```python
df['due_date'] = pd.to_datetime(df['due_date'])
df
```

```python
#Converting the paid_off_time column to pandas datetime format
df['paid_off_time'] = pd.to_datetime(df['paid_off_time'], format='%m/%d/%Y %H:%M')
df
```

```python
#Calculate loan duration
df['loan_duration'] = (df['paid_off_time']-df['effective_date']).dt.days
df
```

```python
#Average time to pay off by education level
avgtime_to_payoff_by_education = df.groupby('education')['loan_duration'].mean()
avgtime_to_payoff_by_education
```

```python
# median principal by term duration
avg_principal_by_term = df.groupby('terms')['Principal'].median()
avg_principal_by_term
            
```

```python
# Percentage of loans paid off early in collections
collection_paidoff_early_percentage = (
    df[(df['loan_status'] == 'collection_paidoff') & (df['paid_off_time'] < df['due_date'])].shape[0] /
    df[df['loan_status'] == 'collection'].shape[0] * 100
)

print(collection_paidoff_early_percentage)

```
